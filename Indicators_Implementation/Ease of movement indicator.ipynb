```python
import pandas as pd
import numpy as np
import yfinance as yf

def calculate_ease_of_movement(df, period=20):
    
    distance_moved = ((df['High'] + df['Low']) / 2) - ((df['High'].shift(1) + df['Low'].shift(1)) / 2)
    
    box_ratio = (df['Volume'] / 100000000) / ((df['High'] - df['Low']) / 1000)
    
    eom = distance_moved / box_ratio
    
    eom_smoothed = eom.rolling(window=period).mean()
    
    df['distance_moved']=distance_moved
    df['box_ratio']=box_ratio
    df['EOM'] = eom_smoothed
    
    
    return df




```

```python
symbol = 'ADANIENT.NS'
start_date = '2021-01-01'
end_date = '2023-01-01'

data = yf.download(symbol, start=start_date, end=end_date)

print(data)

```

```python

eom_data= calculate_ease_of_movement(data)
print(eom_data)

```
